- **count:** Representa el número de valores no nulos en cada columna. En este caso, todos los valores tienen una cuenta de 252, lo que indica que no hay valores faltantes en ninguno de los conjuntos de datos.

- **mean:** Es la media o promedio de los valores en cada columna. Por ejemplo, la media del área de todos los países o territorios es aproximadamente 595,288 kilómetros cuadrados, y la media de la población es aproximadamente 27,276,790.

- **std:** Es la desviación estándar de los valores en cada columna. La desviación estándar mide cuánto se alejan los valores individuales del valor promedio. Por ejemplo, una desviación estándar alta en la columna "population" indicaría una mayor variabilidad en las poblaciones de los países.

- **min:** Representa el valor mínimo en cada columna. Por ejemplo, el valor mínimo en la columna "area" es 0, lo que podría indicar que hay países o territorios con área igual a cero en el conjunto de datos.

- **25%, 50%, y 75%:** Son los percentiles del conjunto de datos. El percentil 25% (Q1) representa el valor por debajo del cual se encuentra el 25% de los datos. El percentil 50% (Q2) es lo mismo que la mediana y representa el valor por debajo del cual se encuentra el 50% de los datos. El percentil 75% (Q3) es el valor por debajo del cual se encuentra el 75% de los datos. Estos percentiles son útiles para entender la distribución de los datos.

- **max:** Es el valor máximo en cada columna. Por ejemplo, el valor máximo en la columna "geoname_id" es aproximadamente 8,505,033.

## Etapa de interpretación
Ahora detectamos los valores atipicos y definimos el limite superior en inferior y desviación estandar para definir la tolerancia.  


```python
anomalies = []

# Funcion ejemplo para detección de outliers
def find_anomalies(data):
    # Set upper and lower limit to 2 standard deviation
    data_std = data.std()
    data_mean = data.mean()
    anomaly_cut_off = data_std * 2
    lower_limit  = data_mean - anomaly_cut_off
    upper_limit = data_mean + anomaly_cut_off
    print(lower_limit.iloc[0])
    print(upper_limit.iloc[0])

    # Generate outliers
    for index, row in data.iterrows():
        outlier = row # # obtener primer columna
        # print(outlier)
        if (outlier.iloc[0] > upper_limit.iloc[0]) or (outlier.iloc[0] < lower_limit.iloc[0]):
            anomalies.append(index)
    return anomalies

find_anomalies(df_espanol.set_index('alpha_3')[['population']])
```

Cuando se detecta un outlier, como Brasil y USA en este caso, se considera que su valor de población estaba muy lejos del rango típico de valores en el conjunto de datos. Esto podría indicar que había algún factor inusual o excepcional en la población de estos países en el momento de la recolección de datos.

```python
# Quitemos BRA y USA por ser outlies y volvamos a graficar:
df_espanol.drop([30,233], inplace=True)
df_espanol.set_index('alpha_3')[['population','area']].sort_values(["population"]).plot(kind='bar',rot=65,figsize=(20,10))
```
